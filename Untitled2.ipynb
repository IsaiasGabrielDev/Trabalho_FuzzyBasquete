```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy.control as ctrl
import skfuzzy as fuzz
```

```python
    if __name__=='__main__':
            print("exemplo")

            #variaveis de entrada 
            distancia_marcador = ctrl.Antecedent(np.arange(0,3,0.10),'distancia_marcador')
            altura_marcador = ctrl.Antecedent(np.arange(0,3,0.10),'altura_marcador' )

            #variaveis de saida
            risco = ctrl.Consequent(np.arange(0,101,1),'risco')

            #variaveis/ fuzzyficaçao 
            distancia_marcador['baixa'] = fuzz.trapmf(distancia_marcador.universe,[0,0,0,1.50])
            distancia_marcador['media'] = fuzz.trimf(distancia_marcador.universe,[1,1.5,2])
            distancia_marcador['alta'] = fuzz.trapmf(distancia_marcador.universe,[1.5,3,3,3])

            altura_marcador ['baixo'] = fuzz.trapmf(altura_marcador.universe,[0,0,0,1.85])
            altura_marcador ['alto'] = fuzz.trapmf(altura_marcador.universe,[1.9,2.2,2.3,2.3])
            
            risco['alto'] = fuzz.trapmf(risco.universe,[0,0,40,60])
            risco['medio'] = fuzz.trimf(risco.universe,[40,60,80])
            risco['pequeno'] = fuzz.trapmf(risco.universe,[60,80,100,100])

            #print os traços 
            distancia_marcador.view()
            altura_marcador.view()
            risco.view()
            plt.show()
```

```python
          #definicao das regras 

          regra1 = ctrl.Rule(distancia_marcador['alta'] & altura_marcador['baixo'],risco['pequeno'])
          regra2 = ctrl.Rule(distancia_marcador['alta'] & altura_marcador['alto'],risco['medio'])
          regra3 = ctrl.Rule(distancia_marcador['media'] & altura_marcador['alto'],risco['medio'])
          regra4 = ctrl.Rule(distancia_marcador['media'] & altura_marcador['baixo'],risco['medio'])
          regra5 = ctrl.Rule(distancia_marcador['baixa'] & altura_marcador['baixo'],risco['pequeno'])
          regra6 = ctrl.Rule(distancia_marcador['baixa'] & altura_marcador['alto'],risco['alto'])

          #ativacao das regras 

          controle_risco = ctrl.ControlSystem([regra1,regra2,regra3,regra4, regra5, regra6])
          simulador_risco = ctrl.ControlSystemSimulation(controle_risco)

          simulador_risco.input['distancia_marcador']= 1
          simulador_risco.input['altura_marcador']= 2.10

          simulador_risco.compute()

          #agregação

          print(simulador_risco.output['risco'])
          risco.view(sim=simulador_risco)
          plt.show()
```
